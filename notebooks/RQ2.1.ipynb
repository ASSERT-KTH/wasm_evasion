```python
import numpy as np

def get_random_search_info(f, alldf, ground):
    # return min, max, medium,  number of success runs (with zero detection at the end), number of runs
    
    finished = {
        'wasm': [],
        'seed': [],
        'finished_len': [],
        'type': [],
        'initial_detectors': []
    }
    
    ground_row = ground[ground['id'] == f]
    
    increased_detection = 0
    total_count_of_rows = 0
    breaking_oracle = 0
    pass_ = 0
    non_pass = 0
    
    for i, d in enumerate(alldf):
        d = d.sort_values("date")
        #d = d[d['engines'] >= 58]
        #display(d)
        total_count_of_rows += len(d)
        breaking_oracle += len(d[d['non_benign'] == 0])
        vals = d.iloc[-1]['non_benign']
        #print(vals)
        if vals == 0:
            finished['wasm'].append(f)
            finished['seed'].append(i)
            finished['finished_len'].append(len(d))
            finished['type'].append('pass')
            pass_ += 1
        else:
            finished['wasm'].append(f)
            finished['seed'].append(i)
            finished['finished_len'].append(len(d))
            finished['type'].append("non_pass")
            
            non_pass += 1
            
        # Get this from the ground dataset
        # display(ground_row)
        finished['initial_detectors'].append(ground_row.iloc[0]['non_benign'])
        # Check if at some point we have more detected than initial
        for row in d.iloc:
            if row['non_benign'] > ground_row.iloc[0]['non_benign']:
                #print(f, i, row['id'])
                # locate these ones
                increased_detection += 1
                # TODO, save all the binaries as a database, this could help us for time evaluation
                # TODO print the detection engines, it can be just random noise
        
        
    
    
    info = pd.DataFrame(finished)
    
    if len(info[info['type'] == 'pass']) > 0:
        print(f, "Breaking with", len(info[info['type'] == 'pass']), "seeds")
    
        sumup = pd.DataFrame({
            'hash': [info.iloc[0]['wasm']],
            'initial_detectors': [ground_row.iloc[0]['non_benign']],
            'max': [max(info[info['type'] == 'pass']['finished_len'].values)],
            'min': [min(info[info['type'] == 'pass']['finished_len'].values)],
            'avg': [np.median(info[info['type'] == 'pass']['finished_len'].values)],
            'mean': [np.mean(info[info['type'] == 'pass']['finished_len'].values)],
            'increased_detection': [increased_detection],
            'generated_files': [total_count_of_rows],
            'breaking_oracle': [breaking_oracle],
            'pass': [pass_],
            'non_pass': [non_pass],
            'ratio': [len(info[info['type'] == 'pass'])/(len(alldf))],
        })
    else:
        print(f, "Non breaking")
        sumup = pd.DataFrame({
            'hash': [info.iloc[0]['wasm']],
            'initial_detectors': [ground_row.iloc[0]['non_benign']],
            'increased_detection': [increased_detection],
            'generated_files': [total_count_of_rows],
            'breaking_oracle': [breaking_oracle],
            'pass': [pass_],
            'non_pass': [non_pass],
            'ratio': [0],
        })
        
    return info, sumup
        
all_ = pd.DataFrame()
for f, alldfs in progress:
    print(f)
    
    try:
        info, suma = get_random_search_info(f, alldfs, vt_benchmark)
        all_ = pd.concat([all_, suma], axis=0)
    except:
        pass

#all_ = all_.reset_index()
all_.reset_index(drop=True, inplace=True)


all_["hash"] = all_["hash"].map(lambda x: x[:8])
display(all_)
all_.to_csv("RQ2.csv")

```
