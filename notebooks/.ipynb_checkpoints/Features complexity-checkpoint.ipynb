```python
import random
random.seed(0)

# Select 2 features from each group
# Select based on prob weight
weights = low['mutability on the wild'].values
d1 = low.sample(n=2, replace=False, weights=weights)


# Remove one of the adds
#alldf = alldf[alldf['feature name'] != 'remove_item_table']
middle = middle[middle['feature name'] != 'add_type']
middle = middle[middle['feature name'] != 'code_motion_ifs']

weights = middle['mutability on the wild'].values
d2 = middle.sample(n=2, replace=False, weights=weights)


weights = high['mutability on the wild'].values
d3 = high.sample(n=2, replace=False, weights=weights)


weights = high2['mutability on the wild'].values
d4 = high2.sample(n=2, replace=False, weights=weights)


weights = high3['mutability on the wild'].values
d5 = high3.sample(n=2, replace=False, weights=weights)

# Have some as hard included
d6 = df[(df['feature name'] == 'peep.select-invert') | (df['feature name'] == 'remove_item_global')]

# concat all features to create the swarm testing
alldf =pd.concat([d1, d2, d3, d4, d5, d6], axis=0)

display(alldf)

# Just to see the distribution per weight

import matplotlib.pyplot as plt

plt.hist(alldf['Naïve weight'].values)
```

```python
# Create the swarm testing
def all_combinations(features, r, cb):
    
    if len(features) == 0:
        cb(r)
    else:
        l = features[0]
        tail = [x for x in features[1:]]
        all_combinations(tail, r + [l], cb)
        all_combinations(tail, r, cb)
        
def get_uniform(features, count):
    
    r = []
    sets = []
    alla_filtered = [r for r in features]
    
    print("Set size", len(alla_filtered))
    
    if count > 2**(len(alla_filtered)):
        # Filter ouur, there is not possible to get more than this number
        print("Setting the count to the max possible")
        all_combinations(alla_filtered, [], lambda x: r.append([ y for y in x ]))
        return r
        
    
    while len(r) < count:
        alla = [r for r in alla_filtered]
        #print(alla)
        selected = [] 
        
        for x in alla:
            n = random.random()
            if n <= 1/len(features):
                selected.append(x)
                
        selected = list(sorted(selected))
        if len(selected) > 0 and f"{selected}" not in sets:
            r.append(selected)
            sets.append(f"{selected}")
    
    return r
```

```python
features = get_uniform(alldf['feature name'].values, 5000)
print(features)
```

```python

def print_to_compile_for_argo(cfgs, create_script=True, attempts=1000, pref=""):
    
    for i, features in enumerate(cfgs):
    
        #print(features)
        S = ""
        
        has_peep = False
        has_remo_item = False
        has_codemotion = False
        modify_custom_section = False
        
        for name in features:
            if "peep" in name:
                has_peep = True
            if "remove_item" in name:
                has_remo_item = True
            if "codemotion" in name:
                has_codemotion = True
            if "code_motion" in name:
                has_codemotion = True
            if "modify_custom_section" in name:
                modify_custom_section = True
            
            S += f"wasm-mutate/{name},"
        if has_peep:
            S += f"wasm-mutate/peep_hole,"
        if has_remo_item:
            S += f"wasm-mutate/remove_item,"
        if has_codemotion:
            S += f"wasm-mutate/code_motion,"
        if modify_custom_section:
            S += f"wasm-mutate/modify_custom_section,"
        
        if create_script:
            #print("cargo build --release --features $FEATURES")
            #print(f"RUST_LOG=debug ./target/release/analyzer --dbconn datas/cf$C mutate -s 100 -e --attempts {c}  --input $INPUT --oracle $@ 2>$C.logs.txt 1> $C.stats.txt")
            print(f"- {{ dbconn: \"c{i}{pref}\", features: \"{S[:-1]}\" }}")
            
        
        
```

```python
print_to_compile_for_argo(features, attempts=1000, pref="vt")
```
