```python

import os
import subprocess
import json

def get_diff(w1, w2, check_folders=["../RQ1/filtered-binaries-metadata_2/pairwise2"]):
    w1 = w1.replace(".wasm", "")
    w2 = w2.replace(".wasm", "")
    
    for check_folder in check_folders:
        for path , dirs, files in os.walk(check_folder):
            for f in files:
                name1 = f"{w1}.wasm.wat_{w2}.wasm.wat.txt"
                name2 = f"{w2}.wasm.wat_{w1}.wasm.wat.txt"
                if name1 == f or name2 == f:
                    return open(f"{path}/{f}", "r").read()
        

def get_debug_info(wasm, check_folders=["../RQ1/filtered-binaries-metadata_2/malware","../tmp_bck2/upload"], 
                   checkmetafile="../RQ1/filtered-binaries-metadata_2/filtered.pretty.json"):
    meta = json.loads(open(checkmetafile, "r").read())
    
    for check_folder in check_folders:
        for path, dirs, files in os.walk(check_folder):
            for f in files:
                if f"{wasm}.wasm" == f:
                    metadata = meta[wasm]
                    
                    numberoffiles = len(metadata['files'])
                    # check the number of instructions per file
                    
                    #print(path, f, numberoffiles, metadata)
                    collections = []
                    
                    number_of_instructions = metadata['instruction_count']
                    size_bytes = metadata['size_bytes']
                    
                    for wasmfile in metadata['files']:
                        collectionmethod = wasmfile['collection_method']
                        
                        collections.append(collectionmethod)
                        
                    return collections, number_of_instructions, size_bytes
                    
    
def get_site(wasm, check_folder="../selenium_scripts/wasms"):
    
    filesin = os.listdir(check_folder)
    filesin = [f for f in filesin if f.endswith("meta.txt")]
    
    fs = []
    
    for f in filesin:
        # read file
        content = open(f"{check_folder}/{f}", "r")
        content = content.read()
        if wasm.replace(".wasm", "").replace("koko_", "") in content:
            fs.append(f.replace(".txt", ""))
            
    return fs
```

```python
def dfs(nodes, clusters, labels):
    
    visited = []
    groups = {}
    
    def dfs_visit(n, clusters, labels, cl):
        if cl not in groups:
            groups[cl] = []
        
        groups[cl].append(n)
        visited.append(n)
        
        # get all connected nodes
        for c in clusters:
            nodes = clusters[c]
            if n in nodes:
                for n2 in nodes:
                    if n != n2:
                        if n2 not in visited:
                            dfs_visit(n2, clusters, labels, cl)
    cl = 0
    for n in nodes:
        if n not in visited:
            dfs_visit(n, clusters, labels, cl)
            cl += 1
            
    return groups
    
#print(clusters)
groups = dfs(list(range(len(nodes))), clusters, labels)

T = 0
GC = 0
T2 = 0
clusterframes = []

for g in groups:
    #print("SAME")
    C = 0
    s = set()
    GC += 1
    
    latextable = {
        'source': [],
        'producer': [],
        'size': [],
        'no. instructions': [],
        'no. functions': [],
        'no. detectors': [],
        'live deploy': [],
        'reproducible': [],
        #'exact': []
    }

    sort = sorted(groups[g], key= lambda x: nodes[x], reverse=True)
    repro = False
    reproname = []
    koko = None
    # Do a pairwise to analyze their differences
    
    header = {
        'hash': []
    }
    last = None
    for n in sort:
        
        filt = all_[all_['wasm'] == f"{nodes[n]}.wasm"]
        if len(filt) > 0:
            row1 = filt.iloc[0]
            header[f"{row1['wasm']}"] = []
            last = f"{row1['wasm']}"
            header['hash'].append(row1['wasm'])
            for n2 in sort:
                #if n2 != n:
                filt = all_[all_['wasm'] == f"{nodes[n2]}.wasm"]

                if len(filt) > 0:
                    row2 = filt.iloc[0]

                    if n != n2:
                        # only in top down
                        #print(row1['wasm'], row2['wasm'])
                        d = get_diff(row1['wasm'], row2['wasm'])
                        #print("=====================================================================")
                        #print(d)
                        if f"{row2['wasm']}" not in header:
                            header[f"{row1['wasm']}"].append(d)
                        else:
                            header[f"{row1['wasm']}"].append("")
                            
                        
                    else:
                        header[f"{row1['wasm']}"].append("")
                        
                    
    # remove last column in heade
    if last:
        del header[last]
        
    headerfr = pd.DataFrame(header)
    # display(headerfr)
    
    headerfr = headerfr.set_index('hash')
    
    p = pd.DataFrame(headerfr).style.to_latex()  

    # patch for latex
    p = p.replace("_", "\\_")
    p = p.replace("$", "\\$")
    
    #print(p)
                
    latextable = {
        'hash': [],
        #**header,
        **latextable
    }        
    
    for n in sort:
        
        #print(nodes[n])
        #print("\t", nodes[n]) #, info[n])
        
        # get from the table
        filt = all_[all_['wasm'] == f"{nodes[n]}.wasm"]
        repro = "koko" in nodes[n] or repro
        if "koko" in nodes[n]:
            koko = nodes[n]
        if repro:
            reproname += get_site(koko)
        if len(filt) > 0:
            row = filt.iloc[0]
            #print(row['wasm'], row['initial_detectors'])
            debug_info = get_debug_info(nodes[n])
            collections, instructions, size = debug_info
            
            latextable['hash'].append(row['wasm'])
            latextable['source'].append(", ".join(list(set(collections)))) # get producer from WASM2WAT parsing
            latextable['producer'].append('todo') # get producer from WASM2WAT parsing
            latextable['size'].append(size) # read from filesystem
            latextable['no. instructions'].append(instructions) # read from WASM2WAT pars
            latextable['no. functions'].append('todo') # read from WASM2WAT parsinging
            latextable['no. detectors'].append(row['initial_detectors'])
            latextable['live deploy'].append(", ".join(list(set(reproname))))
            latextable['reproducible'].append(repro) # if koko prefix
            #if koko:
            #    latextable['exact'].append(koko.replace("koko_", "") == nodes[n]) # if koko prefix
            #else:
            #    latextable['exact'].append(False) # if koko prefix
            
        else:
            print(nodes[n])
            # TODO, ask vt proxy for this
    
        s.add(nodes[n])
        
    display(latextable)
    T += 1
            
            
        
    
    # One table per cluster
    frame = pd.DataFrame(latextable)
    frame = frame.sort_values('no. detectors', ascending=False)

    if len(frame) > 0:
        #frame = frame.set_index('hash')
        #display(frame)
        frame.to_csv(f"{n}.csv")
        p = pd.DataFrame(frame).style.to_latex()  
        
        # patch for latex
        p = p.replace("_", "\\_")
        p = p.replace("$", "\\$")
        T2 += 1
        clusterframes.append(frame)
        #print(p)
        # print head as representative
    #print(len(s), C)
    
    
print(T2)
print(GC)
```

```python
def plot_heat(labelsx, labelsy, mp, saveto, title=""):
    plt.rcParams['figure.figsize'] = [5, 5]

    labelsx = [x[:8] for x in labelsx]
    labelsy = [x[:8] for x in labelsy]

    fig, ax = plt.subplots()
    im = ax.imshow(mp)

    # Show all ticks and label them with the respective list entries
    ax.set_xticks(np.arange(len(labelsx)), labels=labelsx, fontsize=16)
    ax.set_yticks(np.arange(len(labelsy)), labels=labelsy, fontsize=16)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    for i in range(len(labelsx)):
        for j in range(len(labelsy)):
            text = ax.text(j, i, f"{mp[i][j]:.2f}",
                           ha="center", va="center",  fontsize=12)

    if title:
        ax.set_title(title, fontsize=17)
    fig.tight_layout()
    plt.savefig(f"plots/{saveto}.pdf")
    plt.show()
    
```

```python
import re

def load_diff(w1, w2, folder_to_check=["../RQ1/filtered-binaries-metadata_2/pairwise4"]):
    w1 = w1.replace(".wasm", "")
    w2 = w2.replace(".wasm", "")
    
    for check_folder in folder_to_check:
        for path , dirs, files in os.walk(check_folder):
            for f in files:
                name1 = f"{w1}.wasm.wat_{w2}.wasm.wat.txt"
                name2 = f"{w2}.wasm.wat_{w1}.wasm.wat.txt"
                if name1 == f or name2 == f:
                    # get all Js
                    content = open(f"{path}/{f}", "r").read()
                    r = []
                    # JTpe
                    r.append(float(re.findall("TYPES JACCARD DISTANCE: (\d+\.\d+)", content)[0]))
                    
                    # JI
                    r.append(float(re.findall("IMPORTS JACCARD DISTANCE: (\d+\.\d+)", content)[0]))
                    
                    # JE
                    r.append(float(re.findall("EXPORTS JACCARD DISTANCE: (\d+\.\d+)", content)[0]))
                    
                    # JG
                    r.append(float(re.findall("GLOBALS JACCARD DISTANCE: (\d+\.\d+)", content)[0]))
                    
                    # JT
                    r.append(float(re.findall("TABLES JACCARD DISTANCE: (\d+\.\d+)", content)[0]))
                    
                    # JD
                    r.append(float(re.findall("DATA JACCARD DISTANCE: (\d+\.\d+)", content)[0]))
                    
                    # JEl
                    r.append(float(re.findall("ELEMS JACCARD DISTANCE: (\d+\.\d+)", content)[0]))
                    
                    # JF
                    r.append(float(re.findall("\tJACCARD DISTANCE: (\d+\.\d+)", content)[0]))
                    
                    return r
    
    return [0]*8            
                
```

```python

def plot_diff(clusterframes, index = 0, titles=[""], id="hash"):
    for f in clusterframes:
        if len(f) > 1:
            diffs = []
            labelsx = []
            labelsy = []
            for w1 in f.iloc:
                f1 = w1[id]
                labelsx.append(f1)
                labelsy.append(f1)
                diffrow = []
                for w2 in f.iloc:
                    f2 = w2[id]
                    diff = load_diff(f1, f2)
                    diffrow.append(diff[index])
                diffs.append(diffrow)

            print(diffs)
        # create a heatmap
            plot_heat(labelsx, labelsy, diffs, f"{labelsx[0]}.{index}", title=titles[index])
        
print("JTpe")
plot_diff(clusterframes,0, ["JTpe", "JI", "JE", "JG", "JT", "JD", "JEl", "JF"])
print("JI")
plot_diff(clusterframes,1, ["JTpe", "JI", "JE", "JG", "JT", "JD", "JEl", "JF"])
print("JE")
plot_diff(clusterframes,2, ["JTpe", "JI", "JE", "JG", "JT", "JD", "JEl", "JF"])
print("JG")
plot_diff(clusterframes,3, ["JTpe", "JI", "JE", "JG", "JT", "JD", "JEl", "JF"])
print("JT")
plot_diff(clusterframes,4, ["JTpe", "JI", "JE", "JG", "JT", "JD", "JEl", "JF"])
print("JD")
plot_diff(clusterframes,5, ["JTpe", "JI", "JE", "JG", "JT", "JD", "JEl", "JF"])
print("JEl")
plot_diff(clusterframes,6, ["JTpe", "JI", "JE", "JG", "JT", "JD", "JEl", "JF"]) 
print("JF")
plot_diff(clusterframes,7, ["JTpe", "JI", "JE", "JG", "JT", "JD", "JEl", "JF"])    
```
