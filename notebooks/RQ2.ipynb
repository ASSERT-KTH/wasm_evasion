```python
import pandas as pd


# For type checking
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
from pandas.api.types import is_object_dtype
```

```python
vt_benchmark = pd.read_csv("vt.csv")
```

```python
display(vt_benchmark)
```

```python
mal = vt_benchmark[vt_benchmark['non_benign'] > 0]

terms = [f for f in list(mal.values.flatten()) if f != 'undetected' and 
         f != 'unable_to_process_file_type' 
         and not f.__str__().endswith(".wasm")
         and f != 'no_response'
         and f != 'timeout'
        ]

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [30, 20]

terms = { t: len([f for f in terms if f == t]) for t in terms }
print(terms)

plt.bar(x=list(range(len(terms.keys()))), height=list(terms.values()))
plt.xticks(list(range(len(terms.keys()))),list(terms.keys()), rotation=45, ha='right')
```
